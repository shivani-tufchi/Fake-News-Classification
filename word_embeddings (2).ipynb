```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
nltk.download('omw-1.4')
import gensim
nltk.download('wordnet')
!pip install word2vec
import word2vec
```

```python
# word tokenization
word_token = []
for i in df["news content"]:
    words = word_tokenize(i)
    word_token.append(words)
```

```python
total_word_token = []
for i in word_token:
  for j in i:
    total_word_token.append(i)
```

```python
print(len(total_word_token))
```

```python
total_sent_token = []
for i in sent_token:
  for j in i:
    total_sent_token.append(i)
```

```python
print(len(total_sent_token))
```
